```python
import random
import numpy as np
import pandas as pd

def generate(num):
    data = [['One', 'Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine']]
    for i in range(num):
        entry = []
        drink = random.randint(1,4)
        entry.append(drink)
        
        vacation = random.randint(1,4)
        entry.append(vacation)
        
        pet = random.randint(1,2)
        entry.append(pet)
        
        school = random.randint(1,2)
        entry.append(school)
        
        decide = random.randint(1,2)
        entry.append(decide)
        
        party = random.randint(1,4)
        entry.append(party)
        
        skill = random.randint(1,3)
        entry.append(skill)
        
        bus = random.randint(1,4)
        entry.append(bus)
        
        consult = random.randint(0,1)
        entry.append(consult)
        
        data.append(entry)
    return data
        
```

```python
data = generate(8)
df = pd.DataFrame(data)
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
for q in df.columns:
    print(df[q].value_counts(normalize = True))
```
